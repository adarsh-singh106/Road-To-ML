# Pearson Correlation: `corr` and `p`  Quick Notes

## 1. `corr` (Correlation Coefficient)
- Measures **strength and direction** of a **linear relationship**
- Range: **-1 to +1**

### Interpretation
| corr value | Meaning |
|-----------|--------|
| +1.0 | Perfect positive linear relationship |
| -1.0 | Perfect negative linear relationship |
| 0.7 to 0.9 | Strong relationship |
| 0.3 to 0.7 | Moderate relationship |
| 0.0 to 0.3 | Weak / no relationship |

### Key Points
- `corr > 0`  as feature increases, label increases
- `corr < 0`  as feature increases, label decreases
- `corr  0`  no linear relationship
- Correlation **does NOT imply causation**

---

## 2. `p` (p-value)
- Measures **statistical significance** of the correlation
- Answers: *Is this correlation real or due to random chance?*

### Interpretation
| p-value | Meaning |
|--------|--------|
| < 0.05 | Statistically significant  |
|  0.05 | Not statistically significant  |

### Key Points
- Low `p`  strong evidence correlation is real
- High `p`  correlation may be accidental
- Always interpret `corr` **together with** `p`

---

## 3. `pearsonr()` Function
```python
from scipy.stats import pearsonr
corr, p = pearsonr(feature_values, label_values)


````python
selected_features = ['age', 'is_female', 'bmi', 'children', 'is_smoking', 'charges',
       'region_northwest', 'region_southeast', 'region_southwest',
       'bmi_category_normal', 'bmi_category_overweight', 'bmi_category_obese']
````

````python
from scipy.stats import pearsonr

results = []
for feature in selected_features:
    corr, p = pearsonr(df_cleaned[feature], df_cleaned['charges'])
    results.append([feature, corr, p])

df_corr = pd.DataFrame(
    results, columns=['Feature', 'Correlation', 'p_value']
)
````

````python
df_corr['abs_corr'] = df_corr['Correlation'].abs()
df_corr.sort_values(by='abs_corr', ascending=False)
````

````python
categorical_features = [
    col for col in df_cleaned.columns
    if col.startswith(('is_', 'region_', 'bmi_category_'))
]
````

````python
from sklearn.feature_selection import f_regression

X = df_cleaned[categorical_features]
y = df_cleaned['charges']

f_stats, p_values = f_regression(X, y)

anova_df = pd.DataFrame({
    'Feature': categorical_features,
    'F_statistic': f_stats,
    'p_value': p_values
}).sort_values(by='p_value')

anova_df
````

````python
alpha = 0.05

anova_df['Decision'] = anova_df['p_value'].apply(
    lambda p: 'Keep Feature' if p < alpha else 'Drop Feature'
)

anova_df
````

````python
keep_features = anova_df[
    anova_df['Decision'] == 'Keep Feature'
]['Feature'].tolist()

drop_features = anova_df[
    anova_df['Decision'] == 'Drop Feature'
]['Feature'].tolist()

print("KEEP FEATURES:", keep_features)
print("DROP FEATURES:", drop_features)
````

````python
mandatory_features = ['age', 'bmi', 'children']
target = ['charges']
````

````python
final_features = mandatory_features + keep_features + target

# Remove duplicates (safe)
final_features = list(dict.fromkeys(final_features))

final_features
````

````python
final_df = df_cleaned[final_features]

final_df.head()
````

````python
df_cleaned['charges_cat'] = pd.qcut(
    df_cleaned['charges'],
    q=4,            # 3 bins = Low / Medium / High
    labels=False
)
````

````python
# for single feature
from scipy.stats import chi2_contingency

table = pd.crosstab(df_cleaned['is_smoking'], df_cleaned['charges_cat'])
chi2, p, _, _ = chi2_contingency(table)

print("Chi-square:", chi2)
print("p-value:", p)
````

````python
# For All features in list 
from scipy.stats import chi2_contingency
import pandas as pd

alpha = 0.05
results = []

for col in categorical_features:
    table = pd.crosstab(df_cleaned[col], df_cleaned['charges_cat'])
    chi2, p, _, _ = chi2_contingency(table)

    results.append({
        'Feature': col,
        'Chi2_stat': chi2,
        'p_value': p,
        'Decision': 'Keep' if p < alpha else 'Drop'
    })

chi_df = pd.DataFrame(results).sort_values(by='p_value')
chi_df # chi_df is changing when bins are chaneged !! Pick Ftest for continous label
````
