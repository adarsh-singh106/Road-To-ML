# Regression 
```text
Regression is a way to predict a number.
It helps us find out how one thing changes when
something else changes.
```

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('student_salary.csv')
````

````python
df.head()
````

````python
len(df.columns)
````

````python
def plotAll(columns,box=(2,2)):
    n = len(columns)
    plt.figure(figsize=(9,6))
    for i in range(1,n+1):
        plt.subplot(box[0],box[1],i)
        sns.histplot(x=df[columns[i-1]],kde=True)
    plt.tight_layout()
plotAll(df.columns)
````

````python
def scatterAll(features, target):
    n = len(features)
    plt.figure(figsize=(8, 4 * n))

    for i, col in enumerate(features):
        plt.subplot(n, 1, i + 1)
        sns.scatterplot(x=df[col], y=df[target])
        plt.xlabel(col)
        plt.ylabel(target)
        plt.title(f"{target} vs {col}")

    plt.tight_layout()
    plt.show()

scatterAll(df.columns[:-1], "Salary_LPA")
````

````python
import numpy as np

def scatterAll(features, target):
    n = len(features)
    plt.figure(figsize=(8, 4 * n))

    for i, col in enumerate(features):
        x = df[col]
        y = df[target]

        # Fit line: y = mx + c
        m, c = np.polyfit(x, y, 1)

        plt.subplot(n, 1, i + 1)
        plt.scatter(x, y, alpha=0.6)
        plt.plot(x, m*x + c, color="red")
        plt.xlabel(col)
        plt.ylabel(target)
        plt.title(f"{target} vs {col}")

    plt.tight_layout()
    plt.show()
scatterAll(df.columns[:-1], "Salary_LPA")
````
